```
from mip import Model, xsum, MAXIMIZE, CONTINUOUS
```

##Dados do sistema

```
nvar = 2
```

##Criação do modelo

```
m = Model('Equipes', sense=MAXIMIZE)
```

##Criação das variáveis do sistema

```
x = [m.add_var(name='x{}'.format(int(i)), var_type=CONTINUOUS, lb=0) for i in range(nvar)]
```

##Criação da função objetivo

```
m.objective = x[0]
```

##Restrições do sistema

```
m += 40*x[1] + 80*(1-x[1]) >= x[0]
m += 80*x[1] + 20*(1-x[1]) >= x[0]
m += x[1] <= 1
```

```
m.optimize()
```

##Resultado

```
print('Optimize Solution: {}'.format([x[i].x for i in range(nvar)]))
```
