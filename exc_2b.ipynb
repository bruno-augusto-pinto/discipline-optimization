```python
from mip import Model, xsum, CONTINUOUS, MINIMIZE
```

##Dados do sistema

```python
nvar = 3
obj_coef = [50, 40, 60]
```

##Criação do modelo

```python
m = Model('Fertilizantes', sense=MINIMIZE)
```

##Criação das variáveis do sistema

```python
x = [m.add_var(name='x{}'.format(int(i)), var_type=CONTINUOUS, lb=0) for i in range(nvar)]
```

##Criação da função objetivo

```python
m.objective = xsum(obj_coef[i] * x[i] for i in range(nvar))
```

##Restrições do sistema

```python
m += 10*x[0] + 45*x[1] + 50*x[2] >= 100000
m += x[0] + x[1] + x[2] >= 12000
m += x[0] + x[1] + x[2] <= 15000
m += x[0] >= 5000
m += x[0] >=0
m += x[1] >=0
m += x[2] >=0
```

```python
m.optimize()
```

##Resultado

```python
print('Optimize Solution: {}'.format([x[i].x for i in range(nvar)]))
```
