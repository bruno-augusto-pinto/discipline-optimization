```
from mip import Model, xsum, MINIMIZE, CONTINUOUS
```

##Dados do sistema

```
nvar = 4
obj_coef = [2, 4, 1.5, 1]

g_coef = [[2, 2, 10, 20], [50, 20, 10, 30], [80, 70, 10, 80]]
g_val = [1, 70, 250]
```

##Criação do modelo

```
m = Model('Dieta', sense=MINIMIZE)
```

##Criação das variáveis do sistema

```
x = [m.add_var(name='x{}'.format(int(i)), var_type=CONTINUOUS, lb=0) for i in range(nvar)]
```

##Criação da função objetivo

```
m.objective = xsum(obj_coef[i] * x[i] for i in range(nvar))
```

##Restrições do sistema

```
for i in range(len(g_val)):
  m += xsum(g_coef[i][j] * x[j] for j in range(nvar)) >= g_val[i]

m += x[0] >=0
m += x[1] >=0
m += x[2] >=0
m += x[3] >=0
```

```
m.optimize()
```

##Resultado

```
print('Optimize Solution: {}'.format([x[i].x for i in range(nvar)]))
```
