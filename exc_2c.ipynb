```python
from mip import Model, xsum, INTEGER, MINIMIZE
```

##Dados do sistema e criação do modelo

```python
cost = [[40, 50, 55, 9999999999999], [9999999999999, 60, 30, 50]]

m = Model('Transporte', sense=MINIMIZE)
```


##Criação das variáveis do sistema

```python
x = [[m.add_var(var_type=INTEGER, lb=0) for j in range(4)] for i in range(2)]
```

##Criação da função objetivo

```python
m.objective = xsum(cost[i][j] * x[i][j] for i in range(2) for j in range(4))
```

##Restrições do sistema

```python
m += x[1][0] == 0
m += x[0][3] == 0

m += xsum(x[0][j] for j in range(4)) == 6000
m += xsum(x[1][j] for j in range(4)) == 9000

m += x[0][0] <= 2500
m += x[1][3] <= 3500

m += xsum(x[i][1] for i in range(2)) <= 4500
m += xsum(x[i][2] for i in range(2)) <= 5500
```

```python
m.optimize()
```

##Resultado

```python
print('Optimize Solution: {}'.format([x[i][j].x for i in range(2) for j in range(4)]))
```
