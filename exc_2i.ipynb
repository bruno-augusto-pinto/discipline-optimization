```
pip install mip
```

```
from mip import Model, xsum, INTEGER, MAXIMIZE
```

##Dados do sistema

```
nvar = 2
obj_coef = [5, 3.5]
```

##Criação do modelo

```
m = Model('Camisas', sense=MAXIMIZE)
```

##Criação das variáveis do sistema

```
x = [m.add_var(name='x{}'.format(int(i)), var_type=INTEGER, lb=0) for i in range(nvar)]
```

##Criação da função objetivo

```
m.objective = xsum(obj_coef[i] * x[i] for i in range(nvar))
```

##Restrições do sistema

```
m += x[0] <= 150
m += x[1] <= 300
m += 1.5*x[0] + x[1]==400
m += x[0] >=0
m += x[1] >=0
```

```
m.optimize()
```

##Resultado

```
print('Optimize Solution: {}'.format([x[i].x for i in range(nvar)]))
```
